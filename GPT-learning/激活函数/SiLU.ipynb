### SiLU 激活函数

公式：

$$
\text{SiLU}(x) = x \cdot \sigma(x)
$$

其中，$\sigma(x)$ 是 Sigmoid 函数，定义为：

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$


```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"

%matplotlib inline
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# 确保绘图在 Jupyter 内嵌
%matplotlib inline

# 生成输入
x = torch.linspace(-5, 5, 500)  # 点数减少到500

# 将张量强制移动到 CPU
x_cpu = x.cpu()

# 1️⃣ ReLU
y_relu = F.relu(x_cpu).cpu()
plt.figure()
plt.plot(x_cpu.numpy(), y_relu.numpy(), color='blue')
plt.title('ReLU')
plt.grid(True)
plt.show()

# 2️⃣ GELU
y_gelu = F.gelu(x_cpu).cpu()
plt.figure()
plt.plot(x_cpu.numpy(), y_gelu.numpy(), color='green')
plt.title('GELU')
plt.grid(True)
plt.show()

# 3️⃣ SiLU / Swish-1
y_silu = F.silu(x_cpu).cpu()
plt.figure()
plt.plot(x_cpu.numpy(), y_silu.numpy(), color='orange')
plt.title('SiLU / Swish-1')
plt.grid(True)
plt.show()

# 4️⃣ Swish β=1
beta = 1.0
y_swish = (x_cpu * torch.sigmoid(beta * x_cpu)).cpu()
plt.figure()
plt.plot(x_cpu.numpy(), y_swish.numpy(), color='red')
plt.title('Swish (β=1)')
plt.grid(True)
plt.show()
```
